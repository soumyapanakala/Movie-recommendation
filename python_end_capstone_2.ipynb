```python
# We want most rated movies.., movie name(title) and user ratings are in different dataframes, merge the dataframes by taking 
# movieId column as common in both datasets

ratingmovie_data=ratings_data[['userId','movieId','rating']].merge(movielinks_data[['movieId','title']],on='movieId',how='left')
ratingmovie_data
```

```python
# Now count the ratings given by users according to title by gouping title
x=ratingmovie_data.groupby(['title'],as_index = False)['userId'].count()
x
```

```python
# Check the titles which have most number of ratings by users by sorting in descending order...

x=x.sort_values("userId",ascending = False)
x = x[:100]          # I have used Slicing to plot the graph easily without clumsy as there are lot of rows...
x
```

```python
# To plot the graph import the tools as per requirement---
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the bar graph of most rated movies

plt.figure(figsize = (40,15))
sns.barplot(x = x['title'], y = x['userId'])
plt.title("Most rated Movies", fontsize = 36, color = 'red')
plt.xlabel("Movie name", fontsize = 32, color = 'blue')
plt.ylabel("Number of ratings",fontsize = 32, color = 'blue')
plt.xticks(rotation = 75, horizontalalignment = 'right',fontsize = 18)
plt.yticks(fontsize = 20)
plt.show()
```

```python
# top 10 users who has rated most
#for this according to userid count number of ratings given by users.. and then sort values from highest to lowest and 
# then take top 10 users

top = ratings_data.groupby(['userId'],as_index = False)['rating'].count().sort_values("rating",ascending = False).head(10)
top
```

```python
# Now plot the graph of top 10 users who rated most

plt.figure(figsize = (10,4))
sns.barplot(x = top['userId'], y = top['rating'], width = 0.5)
plt.title("Most rated users", fontsize = 20, color = 'red')
plt.xlabel("userId's" , fontsize = 16, color = 'blue')
plt.ylabel('Number of ratings', fontsize = 16, color = 'blue')
plt.show()  
```
